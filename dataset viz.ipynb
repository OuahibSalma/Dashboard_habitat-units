```python
# Ce script pour la resolution de probleme de nommage pour les colonnes , on a probleme de ligne et de colonnes 
# Resolution de colonnes 
new_columns_names = { 
    "Unités d'habitat mis en chantier par région et par catégorie 2016-2020" : 'Regions' , 
    'Unnamed: 1' : 'Categories' , 
    'Unnamed: 2' : 'Lots_2016' ,
    'Unnamed: 3' : 'Logements_2016' ,
    'Unnamed: 4' : 'TotalUnites_2016' ,
    'Unnamed: 5' : 'Lots_2017' ,
    'Unnamed: 6' : 'Logements_2017' ,
    'Unnamed: 7' : 'TotalUnites_2017' ,
    'Unnamed: 8' : 'Lots_2018' ,
    'Unnamed: 9' : 'Logements_2018' ,
    'Unnamed: 10' : 'TotalUnites_2018' ,
    'Unnamed: 11' : 'Lots_2019' ,
    'Unnamed: 12' : 'Logements_2019' ,
    'Unnamed: 13' : 'TotalUnites_2019' ,
    'Unnamed: 14' : 'Lots_2020' ,
    'Unnamed: 15' : 'Logements_2020' ,
    'Unnamed: 16' : 'TotalUnites_2020' ,
}
data = data.rename(columns=new_columns_names)
# Supprimer les quatre premières lignes
data = data.drop([0, 1, 2, 3])

# Réinitialiser les index
data.reset_index(drop=True, inplace=True)

data.head(20)
```

```python
data['Regions'].ffill(inplace=True)
data
```
